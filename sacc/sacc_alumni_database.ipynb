 ### CREATE THE DATABASE

```python
import sqlite3

#Create the database
conn = sqlite3.connect('sacc_alumni_database.db')

#Create a cursor object to interact with the database 
cursor = conn.cursor()
```

```python
#Create the alumni table

cursor.execute( '''
CREATE TABLE IF NOT EXISTS alumni(
               id INTEGER PRIMARY KEY AUTOINCREMENT,
               Firstname TEXT NOT NULL,
               Lastname TEXT NOT NULL,
               age INTEGER,
               sex TEXT,
               year_of_completion INTEGER,
               occupation TEXT,
               location TEXT,
               phone_number TEXT,
               email TEXT,
               society TEXT
               )

''')

#commit changes
conn.commit()
```

### CREATE THE UI USING TKINTER

```python
import sqlite3
from tkinter import *
from tkinter import messagebox

# Function to insert data into the SQLite database
def insert_data():
    firstname = fname_var.get()
    lastname = lname_var.get()
    age = age_var.get()
    sex = sex_var.get()
    location = location_var.get()
    year = year_var.get()
    occupation = occupation_var.get()
    phone = phone_var.get()
    email = email_var.get()
    society = society_var.get()

    # Validate that all fields are filled
    if not (firstname and lastname and age and sex and location and year and occupation and phone and email):
        messagebox.showwarning("Input Error", "All fields must be filled But SOCIETY is Optional")
        return

    # Connect to SQLite database and insert data
    conn = sqlite3.connect('sacc_alumni_database.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO alumni (firstname, lastname, age, sex, location, year_of_completion, occupation, phone_number, email, society)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (firstname, lastname, age, sex, location, year, occupation, phone, email, society))

    conn.commit()
    conn.close()

    # Show success message
    messagebox.showinfo("Success", "Data inserted successfully!")
    clear_form()

# Function to clear the form fields after submission
def clear_form():
    fname_var.set("")
    lname_var.set("")
    age_var.set("")
    sex_var.set("")
    location_var.set("")
    year_var.set("")
    occupation_var.set("")
    phone_var.set("")
    email_var.set("")
    society_var.set("")

# Create the main window
root = Tk()
root.title("Church Alumni Data Entry")
root.geometry("400x400")
root.config(bg='#f0f0f0')  # Set the background color of the window

# Define Tkinter variables to store form data
fname_var = StringVar()
lname_var = StringVar()
age_var = StringVar()
sex_var = StringVar()
location_var = StringVar()
year_var = StringVar()
occupation_var = StringVar()
phone_var = StringVar()
email_var = StringVar()
society_var = StringVar()

# Define a common font and style for all labels and entries
label_font = ('Helvetica', 12)
entry_font = ('Helvetica', 12)
bg_color = '#f0f0f0'

# Create a frame to organize form elements using grid
frame = Frame(root, bg=bg_color)
frame.pack(padx=20, pady=20)

# Labels and Entry Fields in Grid Layout
Label(frame, text="First Name", font=label_font, bg=bg_color).grid(row=0, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=fname_var, font=entry_font, width=30).grid(row=0, column=1, padx=10, pady=5)

Label(frame, text="Last Name", font=label_font, bg=bg_color).grid(row=1, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=lname_var, font=entry_font, width=30).grid(row=1, column=1, padx=10, pady=5)

Label(frame, text="Age", font=label_font, bg=bg_color).grid(row=2, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=age_var, font=entry_font, width=30).grid(row=2, column=1, padx=10, pady=5)

Label(frame, text="Sex", font=label_font, bg=bg_color).grid(row=3, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=sex_var, font=entry_font, width=30).grid(row=3, column=1, padx=10, pady=5)

Label(frame, text="Location", font=label_font, bg=bg_color).grid(row=4, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=location_var, font=entry_font, width=30).grid(row=4, column=1, padx=10, pady=5)

Label(frame, text="Year of Completion", font=label_font, bg=bg_color).grid(row=5, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=year_var, font=entry_font, width=30).grid(row=5, column=1, padx=10, pady=5)

Label(frame, text="Occupation", font=label_font, bg=bg_color).grid(row=6, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=occupation_var, font=entry_font, width=30).grid(row=6, column=1, padx=10, pady=5)

Label(frame, text="Contact Number", font=label_font, bg=bg_color).grid(row=7, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=phone_var, font=entry_font, width=30).grid(row=7, column=1, padx=10, pady=5)

Label(frame, text="Email", font=label_font, bg=bg_color).grid(row=8, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=email_var, font=entry_font, width=30).grid(row=8, column=1, padx=10, pady=5)

Label(frame, text="Society(ies)", font=label_font, bg=bg_color).grid(row=9, column=0, padx=10, pady=5, sticky=E)
Entry(frame, textvariable=society_var, font=entry_font, width=30).grid(row=9, column=1, padx=10, pady=5)

# Submit button in a separate row
Button(frame, text="Submit", font=label_font, command=insert_data, bg='#4CAF50', fg='white', width=20).grid(row=11, column=0, columnspan=2, pady=20)

# Run the Tkinter event loop
root.mainloop()
```
